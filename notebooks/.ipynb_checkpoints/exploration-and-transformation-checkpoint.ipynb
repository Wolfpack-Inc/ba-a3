### Drop insignificant variables

````python
# Drop insignificant variables
drop_variables = ['s1', 's10', 's16', 's18', 's19', 's5', 's6', 'setting3']

# Training set
for variable in drop_variables:
    df_training.drop(variable,
                     axis = 1,
                     inplace=True)

# Test set
for variable in drop_variables:
    df_predict.drop(variable,
                    axis = 1,
                    inplace=True)
````

<br><br><br>
# Pre-process data

### Add RUL

````python
%%time

# Add the remaining cycles (RUL)
def countAllCycles(engine_id, df):
    """
    This function calculates all the .
    """
    # Extract amount of cycles
    dfFiltered     = df[df['engine_id'] == engine_id]
    amountOfCycles = len(dfFiltered)
    
    return amountOfCycles

# Add the total amount of cycles
df_training['total_cycles']     = df_training['engine_id'].apply(lambda x: countAllCycles(x, df_training))

# Calculate and add remaining cycles (rul)
df_training['rul'] = df_training['total_cycles'] - (df_training['cycle'] - 1) 

# Drop the 'total_cycles' column
df_training.drop(labels = 'total_cycles', axis = 1, inplace = True)
df_training.head()
````

## Standardize the variables
We use the StandardScaler and fit it first on our df_training set and afterwards use the fitted scaler on the df_training and df_predict set.

````python
# Create a scaler on x, transform x and also apply it on y
featuresWithoutEngineidCycleAndTarget = df_training.columns[2:-1]

x_train = df_training.loc[:, featuresWithoutEngineidCycleAndTarget].values
scaler  = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)

x_test = df_predict.loc[:, featuresWithoutEngineidCycleAndTarget].values
x_test = scaler.transform(x_test)

# Transform the scaled data into dataframes, and overwrite the df_training and df_predict
x_train = pd.DataFrame(data = x_train, columns = featuresWithoutEngineidCycleAndTarget)
x_test  = pd.DataFrame(data = x_test, columns = featuresWithoutEngineidCycleAndTarget)

df_training.update(x_train)
df_predict.update(x_test)
````
